```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('EnergyEfficiency.csv')
```

```python
sns.pairplot(df)
plt.show()
```

```python
cmatrix = df.corr(numeric_only=True)

sns.heatmap(cmatrix, annot=True, cmap='RdYlGn')
plt.title('Correlation Matrix')
plt.show()
```

```python
sns.regplot(data=df, x='Compactness', y='HeatingLoad', line_kws={'color': 'blue'})
plt.xlabel('Compactness')
plt.ylabel('Heating Load')
plt.title('Compactness vs. Heating Load')
plt.show()

sns.regplot(data=df, x='Compactness', y='CoolingLoad', line_kws={'color': 'blue'})
plt.xlabel('Compactness')
plt.ylabel('Cooling Load')
plt.title('Compactness vs. Cooling Load')
plt.show()
```

```python

sns.regplot(data=df, x='WallArea', y='HeatingLoad', line_kws={'color': 'blue'})
plt.xlabel('Wall Area')
plt.ylabel('Heating Load')
plt.title('Wall Area vs. Heating Load')
plt.show()

sns.regplot(data=df, x='WallArea', y='CoolingLoad', line_kws={'color': 'blue'})
plt.xlabel('Wall Area')
plt.ylabel('Cooling Load')
plt.title('Wall Area vs. Cooling Load')
plt.show()
```

```python
sns.regplot(data=df, x='Height', y='HeatingLoad', line_kws={'color': 'blue'})
plt.xlabel('Height')
plt.ylabel('Heating Load')
plt.title('Height vs. Heating Load')
plt.show()

sns.regplot(data=df, x='Height', y='CoolingLoad', line_kws={'color': 'blue'})
plt.xlabel('Height Level')
plt.ylabel('Cooling Load')
plt.title('Height vs. Cooling Load')
plt.show()
```

```python
sns.regplot(data=df, x='GlazingLevel', y='HeatingLoad', line_kws={'color': 'blue'})
plt.xlabel('Glazing Level')
plt.ylabel('Heating Load')
plt.title('Glazing Level vs. Heating Load')
plt.show()

sns.regplot(data=df, x='GlazingLevel', y='CoolingLoad', line_kws={'color': 'blue'})
plt.xlabel('Glazing Level')
plt.ylabel('Cooling Load')
plt.title('Glazing Level vs. Cooling Load')
plt.show()
```

```python
# Fill in missing data first
mapping = {0: 'Uniform', 1: 'North', 2: 'East', 3: 'South', 4: 'West'}

df['GlazingAreaDistribution'] = df['GlazingAreaDistribution'].fillna(df['GlazingAreaDistribution'].map(mapping))

df
```

```python
# To show that the data is not really accurate
sns.countplot(data=df, x='Orientation')
plt.xlabel('Orientation')
plt.ylabel('Count')
plt.title('Distribution of Orientation')
plt.show()

# Bar graph with hue for comparison between two categorical variables
sns.countplot(data=df, x='Orientation', hue='GlazingAreaDistribution')
plt.xlabel('Orientation')
plt.ylabel('Count')
plt.title('Distribution of Orientation with Glazing Area Distribution')
plt.show()

# Data is artificial as everything is equal.
```
