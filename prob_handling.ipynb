```python
import pandas as pd

df = pd.read_csv('EnergyEfficiency.csv')
```

```python
df = df[df['GlazingAreaDistribution'] != 'None']

mapping = {
    'North': 1,
    'East': 2,
    'South': 3,
    'West': 4,
    'Uniform': 5
}

r_mapping = {
    1: 'North',
    2: 'East',
    3: 'South',
    4: 'West',
    5: 'Uniform'
}
# Map values 
df['GlazingAreaDistribution'] = df['GlazingAreaDistribution'].map(mapping)
# U
df['GlazingAreaDistribution'] = df['GlazingAreaDistribution'].fillna(method='ffill')

df['GlazingAreaDistribution'] = df['GlazingAreaDistribution'].map(r_mapping)

df
```
